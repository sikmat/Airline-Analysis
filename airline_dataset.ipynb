```python
# Function to categorize age groups
def cat_age (string):
  if str(string) == 'nan':
      return 'Unknown'
  elif int(string) <= 12:
      return 'Child'
  
  elif 12 <= int(string) <= 17:
     return 'Teenager'
  
  elif 18 <= int(string) <= 59:
     return 'Adult'
  
  elif int(string) > 60:
     return 'Elder'
```

```python
airline['age_group']=airline['age'].apply(cat_age)
# Create new column called 'age_group', invoke cat_age function on age column value to determine the age category using the age values
airline.head()
# Defaults to the first five rows
```

```python
airline['age_group'].value_counts()
# Count passengers by age category
```

```python
age_group_count=airline['age_group'].value_counts()
plt.pie(age_group_count, labels=age_group_count.index, autopct='%1.1f%%')
# Plot pie chart using age_group categories and their count
```

```python
airline.groupby(['age_group', 'flight_status']).size().unstack().plot(kind='bar', stacked=False)
plt.title(f'Count of {'flight_status'} based on {'age_group'}')
plt.xlabel(f'Passenger {'age_group'}')
plt.ylabel('Count')
plt.show()
```
